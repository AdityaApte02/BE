```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv("./Assignment3/iris.csv")
```

```python
data.head()
```

```python
data["Species"].value_counts()
```

```python
data.isnull().sum()
```

```python
data.pop("Id")
```

```python
data
```

```python
x=data.drop("Species", axis="columns")
```

```python
x
```

```python
y=data['Species']
```

```python
y
```

```python
x=preprocessing.StandardScaler().fit_transform(x)
```

```python
x
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.25, random_state=42)
```

```python
X_train
```

```python
knnModel = KNeighborsClassifier(n_neighbors=5)
```

```python
knnModel.fit(X_train,Y_train)
```

```python
Y_pred = knnModel.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
accuracy_score(Y_test,Y_pred)
```

```python
conf_matrix=confusion_matrix(Y_test,Y_pred)
```

```python
import seaborn as sns
```

```python
sns.heatmap(conf_matrix, annot=True, cbar=True)
plt.ylabel("True label")
plt.xlabel("Predicted label")
plt.title("Confusion Matrix")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
data = [[2, 4, "Orange"], [4, 4, "Blue"], [4, 6, "Orange"], [4, 2, "Orange"], [6, 2, "Blue"], [6, 4, "Orange"]]
```

```python
dataset=pd.DataFrame(data, columns=["X","Y", "TargetColor"])
```

```python
dataset
```

```python
X=dataset.drop("TargetColor", axis="columns")
```

```python
X
```

```python
Y=dataset["TargetColor"]
```

```python
Y
```

```python
model=KNeighborsClassifier(n_neighbors=3)
model.fit(X,Y)
```

```python
Y_pred=model.predict([[6,6]])
```

```python
Y_pred
```

```python
conf_matrix=confusion_matrix([["Orange"]],Y_pred)
```

```python
accuracy_score([["Orange"]],Y_pred)
```

```python
sns.heatmap(conf_matrix, annot=True, cbar=True)
```
